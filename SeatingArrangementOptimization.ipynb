#### Calculate cost

```python
def calculate_cost(arrangement):
    cost = likes.loc[arrangement[0]][arrangement[len(arrangement) - 1]]
    for i in range(len(arrangement) - 1):
        cost += likes.loc[arrangement[i]][arrangement[i+1]]
    return cost 
```

#### Genetic Algorithm

```python
def generate_random_state(seed=None):
    state = []
    state_length = len(likes.columns)
    while(len(state) < state_length):
        index = random.randint(0, state_length - 1)
        guest = likes.columns[index]
        if(guest not in state):
            state.append(guest)
    cost = calculate_cost(state)
    return state, cost
```

```python
def generate_random_popluation(population_size, seed=None):
    random.seed(seed)
    population = []
    costs = [] 
    while(len(population) < population_size):
        chromosome, cost = generate_random_state(seed)
        if(chromosome not in population):
            population.append(chromosome)
            costs.append(cost)
    return population, costs
```

```python
def genetic_algorithm(population_size=100, num_generations=1000, mutation_rate=0.1):
    # randomly generate initial population
    population, costs = generate_random_popluation(population_size)
    best_arrangement, best_cost = [], 0
    for i in range(num_generations):
        least_cost_index = int(0.6 * population_size)
        least_cost = sorted(costs, reverse=True)[least_cost_index]
        # population = [i for i in population if costs[np.where(i)[0][0]] >= least_cost]
        pop  = []
        for i in range(len(costs)):
            if(costs[i] >= least_cost):
                pop.append(population[i])

        population = pop
        new_population = []
        new_costs = []
        while len(new_population) < population_size:
            invalid_crossing = True
            while(invalid_crossing):
                x = random_select(population)
                y = random_select(population)
                cross_over_point = choose_cross_over_point(x, y)
                if cross_over_point != -1: # random cross over point was found and valid 
                    invalid_crossing = False 
                    first_child, second_child = cross_over(x, y, cross_over_point) 
                    first_child = mutate(first_child, mutation_rate)
                    second_child = mutate(second_child, mutation_rate)
                    new_population.append(first_child)
                    new_population.append(second_child)
                    first_child_cost = calculate_cost(first_child)
                    second_child_cost = calculate_cost(second_child)
                    new_costs.append(first_child_cost)
                    new_costs.append(second_child_cost)
                    if(first_child_cost > best_cost):
                        best_cost = first_child_cost
                        best_arrangement = first_child
                    if(second_child_cost > best_cost):
                        best_cost = second_child_cost
                        best_arrangement = second_child
        population = new_population
        costs = new_costs
    return best_arrangement, best_cost
```

```python
def random_select(population):
    random_index = random.randint(0, len(population) - 1)
    return population[random_index]
```

```python
def choose_cross_over_point(x, y):
    for i in range(20):
        random_index = random.randint(0, len(x) - 2)
        point_is_valid = True
        for gene in x[:random_index + 1]:
            if gene not in y[:random_index + 1]:
                point_is_valid = False
                break 
        if point_is_valid:
            return random_index

    return -1 # random point was not found
```

```python
def cross_over(x, y, cross_over_point):
    first_child, second_child = [], []
    for i in range(cross_over_point + 1): 
        first_child.append(x[i])
        second_child.append(y[i])
    for i in range(cross_over_point + 1, len(x)):
        first_child.append(y[i])
        second_child.append(x[i])
    return first_child, second_child
```

```python
def mutate(chromosome, mutation_rate):
    mutated_chromosome = copy.deepcopy(chromosome)
    array = np.zeros(100)
    freq_of_1 = int(mutation_rate * array.shape[0])
    for i in range(freq_of_1):
        array[i] = 1
    # probability of getting 1 randomly = mutation rate 
    random_value = array[random.randint(0, array.shape[0] - 1)]
    if(random_value == 1):
        # get two radnom positions in the chromosome
        pos_1 = random.randint(0, len(mutated_chromosome) - 1)
        pos_2 = random.randint(0, len(mutated_chromosome) - 1)
        # swap genes at the two positions 
        temp = mutated_chromosome[pos_1]
        mutated_chromosome[pos_1] = mutated_chromosome[pos_2]
        mutated_chromosome[pos_2] = temp

    return mutated_chromosome
```

#### Hill Climbing Algorithm

```python
def hill_climbing(num_restarts=100):
    restartNo = 0
    best_arrangement, best_cost = [], 0
    while(restartNo < num_restarts):
        # start with a random arrangement
        curr_arrangement, curr_cost = generate_random_state()
        neighbours, neighbours_costs = generate_neighbours(curr_arrangement)
        highest_cost = neighbours_costs.max()
        best_neighbour = neighbours[np.where(neighbours_costs == highest_cost)[0][0]] # the first arrangement with highest cost value
        while(True):
           if(highest_cost <= curr_cost):
               if(curr_cost >= best_cost):
                   best_arrangement, best_cost = curr_arrangement, curr_cost
               restartNo += 1
               break
           else:
               curr_arrangement = best_neighbour
               curr_cost = highest_cost
               neighbours, neighbours_costs = generate_neighbours(curr_arrangement)
               highest_cost = neighbours_costs.max()
               best_neighbour = neighbours[np.where(neighbours_costs == highest_cost)[0][0]]        
    return best_arrangement, best_cost
```

```python
def generate_neighbours(arrangement):
    neighbours = []
    costs = []
    for i in range(len(arrangement) - 1):
        for j in range(i+1, len(arrangement)):
            neighbour = copy.deepcopy(arrangement)
            swap(neighbour, i, j)
            neighbours.append(neighbour)
            costs.append(calculate_cost(neighbour))
    return neighbours, np.array(costs)  
```

```python
def swap(arrangement, i, j):
    temp = arrangement[i]
    arrangement[i] = arrangement[j]
    arrangement[j] = temp
```

#### Simulated Annealing Algorithm

```python
def simulated_annealing(initial_temperature=100000, cooling_rate=0.99, num_iterations=10000):
    temp = initial_temperature
    # Start with a random arrangement
    curr_arrangement, curr_cost = generate_random_state()
    for i in range(num_iterations):
        neighbours, neighbours_costs = generate_neighbours(curr_arrangement)
        # randomly select any neighbour 
        random_val = random.randint(0, len(neighbours) - 1)
        selected_neighbour = neighbours[random_val]
        selected_neighbour_cost = neighbours_costs[random_val]
        delta = selected_neighbour_cost - curr_cost 
        # test improvement
        if delta > 0:
            curr_arrangement = selected_neighbour
            curr_cost = selected_neighbour_cost
        else:
            # accept worse solutions with a probability depending on the current temperature to escape local minima.
            probability = math.exp(delta/temp)
            if (accept_worse(probability)):
                curr_arrangement = selected_neighbour
                curr_cost = selected_neighbour_cost
        temp *= cooling_rate # decrease temperature
    return curr_arrangement, curr_cost
```

```python
def accept_worse(probability):
    array = np.zeros(1000)
    freq = int(probability * array.shape[0])
    for i in range(freq):
        array[i] = 1
    random_index = random.randint(0, array.shape[0] - 1)
    return array[random_index] == 1
```
